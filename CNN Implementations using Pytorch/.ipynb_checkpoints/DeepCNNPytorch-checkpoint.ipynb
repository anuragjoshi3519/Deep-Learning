```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torchvision
from torchvision import transforms
import torch.optim as optim
import math
import copy
```

```python
transform_train = transforms.Compose([
                                      transforms.RandomResizedCrop(224),
                                      transforms.ToTensor(),
                                      transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])

transform_test = transforms.Compose([
                                     transforms.RandomResizedCrop(224),
                                     transforms.ToTensor(),
                                     transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])
```

```python
train_set = torchvision.datasets.CIFAR10(root='Images/',train=True,download=True,transform=transform_train)
test_set = torchvision.datasets.CIFAR10(root='Images/',train=False,download=True,transform=transform_test)
```

```python
def showImg(img,title):
    img = img.numpy()*0.5 + 0.5
    img = np.transpose(img,(1,2,0))
    plt.figure(figsize=(8,1))
    plt.title(title)
    plt.imshow(img)
```

```python
def showBatch(dataLoader):
    images,labels = next(iter(dataLoader))
    img = torchvision.utils.make_grid(images)
    title = ' '.join(str(x) for x in list(labels.numpy()))
    showImg(img,title)  
```

```python
dataLoader = torch.utils.data.dataloader.DataLoader(train_set,batch_size=8)
showBatch(dataLoader)
```

# Training

```python
def evaluation(dataLoader,model):
    total=0
    correct=0
    for images, labels in dataLoader:
        images=images.to(device)
        labels=labels.to(device)
        pred = model(images)
        _,pred_label = torch.max(pred,dim=1)
        total+=labels.shape[0]
        correct+=(pred_label==labels).sum().item()
    # print(total,correct)
    return (correct/total)*100
```

```python
def fit(dataLoader,model,opt,loss_fn,epochs):
    best_model=model.state_dict()
    min_loss=1000
    loss_arr=[]
    loss_epoch_arr=[]
    for i in range(epochs):
        for j,data in enumerate(dataLoader,0):
            images,labels = data
            images=images.to(device)
            labels=labels.to(device)
            pred = model(images)
            loss = loss_fn(pred,labels)
            loss_arr.append(loss.item())

            loss.backward()

            if min_loss>loss.item():
                min_loss=loss.item()
                best_model = copy.deepcopy(model.state_dict())
                print("Min Loss = {}".format(min_loss))

            opt.step()
            opt.zero_grad()

            del images,labels
            torch.cuda.empty_cache()
            if j%100==0:
                print('Epoch#{} %% Iteration-{}/{} :: training loss: {}'.format(i,j,n_iters,loss_arr[-1]))
            loss_epoch_arr.append(loss_arr[-1])
            #print('After {} epoch(s) - train accuracy: {}, test accuracy: {}'.format(i,evaluation(dataLoader,model),evaluation(testLoader,model)))

    plt.title("Training Loss")
    plt.plot(loss_epoch_arr)
    plt.show()
    return best_model
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
BATCH_SIZE = 16
trainLoader = torch.utils.data.dataloader.DataLoader(train_set,batch_size=BATCH_SIZE)
testLoader = torch.utils.data.dataloader.DataLoader(test_set, batch_size=BATCH_SIZE)

vgg16=vgg16.to(device)
opt = torch.optim.Adam(vgg16.parameters())
loss_fn=nn.CrossEntropyLoss()
no_of_epochs=1

n_iters = math.ceil(50000/BATCH_SIZE)

best_model=fit(trainLoader,vgg16,opt,loss_fn,no_of_epochs)
```

# **Transfer Learning**

```python
vgg16_TrainedModel = models.vgg16_bn(pretrained=True)
```

```python
for param in vgg16_TrainedModel.parameters():
    param.requires_grad=False
```

```python
final_in_features = vgg16_TrainedModel.classifier[-1].in_features
vgg16_TrainedModel.classifier[-1]=nn.Linear(final_in_features,10)
print(vgg16_TrainedModel.classifier)
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
BATCH_SIZE = 16
trainLoader = torch.utils.data.dataloader.DataLoader(train_set,batch_size=BATCH_SIZE)
testLoader = torch.utils.data.dataloader.DataLoader(test_set, batch_size=BATCH_SIZE)

vgg16_TrainedModel=vgg16_TrainedModel.to(device)
opt = torch.optim.SGD(vgg16_TrainedModel.parameters(),lr=0.05)
loss_fn=nn.CrossEntropyLoss()
no_of_epochs=2

n_iters = math.ceil(50000/BATCH_SIZE)

best_model=fit(trainLoader,vgg16_TrainedModel,opt,loss_fn,no_of_epochs)
```

```python
vgg16_TrainedModel.load_state_dict(best_model)
print('Train accuracy: {}, Test accuracy: {}'.format(evaluation(trainLoader,vgg16_TrainedModel),evaluation(testLoader,vgg16_TrainedModel)))
```

# **Implementing Inception Net**

```python
inception = models.inception_v3(pretrained=True)
#inception
```

```python
for param in inception.parameters():
    param.requires_grad=False

i_in_features = inception.fc.in_features
inception.fc=nn.Linear(i_in_features,10)
auxLogits_in_features=inception.AuxLogits.fc.in_features
inception.AuxLogits.fc=nn.Linear(auxLogits_in_features,10)

for param in inception.parameters():
    if param.requires_grad:
        print(param.shape)
```

```python
def evaluation_inception(dataLoader,model):
    total=0
    correct=0
    for images, labels in dataLoader:
        images=images.to(device)
        labels=labels.to(device)
        pred,aux_pred = model(images)   ## changes
        _,pred_label = torch.max(pred,dim=1) 
        total+=labels.shape[0]
        correct+=(pred_label==labels).sum().item()
    # print(total,correct)
    return (correct/total)*100

def fit_inception(dataLoader,model,opt,loss_fn,epochs):
    best_model_inception=model.state_dict()
    min_loss=1000
    loss_arr=[]
    loss_epoch_arr=[]
    for i in range(epochs):
        for j,data in enumerate(dataLoader,0):
            images,labels = data
            images=images.to(device)
            labels=labels.to(device)
            pred,aux_pred = model(images)  # changes
            loss = loss_fn(pred,labels) + 0.3*loss_fn(aux_pred,labels)
            loss_arr.append(loss.item())
      
        loss.backward()

        if min_loss>loss.item():
            min_loss=loss.item()
            best_model_inception = copy.deepcopy(model.state_dict())
            print("Min Loss = {}".format(min_loss))

        opt.step()
        opt.zero_grad()

        del images,labels
        torch.cuda.empty_cache()
        if j%100==0:
            print('Epoch#{} %% Iteration-{}/{} :: training loss: {}'.format(i,j,n_iters,loss_arr[-1]))
            loss_epoch_arr.append(loss_arr[-1])
        #print('After {} epoch(s) - train accuracy: {}, test accuracy: {}'.format(i,evaluation_inception(dataLoader,model),evaluation_inception(testLoader,model)))
      
        plt.title("Training Loss")
        plt.plot(loss_epoch_arr)
        plt.show()
        return best_model_inception
```

```python
trainTransform_Inception = transforms.Compose([
                                               transforms.RandomResizedCrop(299),
                                               transforms.ToTensor(),
                                               transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])
testTransform_Inception = transforms.Compose([
                                               transforms.RandomResizedCrop(299),
                                               transforms.ToTensor(),
                                               transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])

train_set=torchvision.datasets.CIFAR10(root='./data',train=True,transform=trainTransform_Inception)
test_set=torchvision.datasets.CIFAR10(root='./data',train=False,transform=testTransform_Inception)

trainLoader = torch.utils.data.dataloader.DataLoader(train_set,batch_size=BATCH_SIZE)
testLoader = torch.utils.data.dataloader.DataLoader(test_set, batch_size=BATCH_SIZE)
```

```python
BATCH_SIZE = 16

inception=inception.to(device)
opt = torch.optim.Adam(inception.parameters())
loss_fn=nn.CrossEntropyLoss()
no_of_epochs=4

n_iters = math.ceil(50000/BATCH_SIZE)

best_model_inception=fit_inception(trainLoader,inception,opt,loss_fn,no_of_epochs)

inception.load_state_dict(best_model_inception)
print('Train accuracy: {}, Test accuracy: {}'.format(evaluation_inception(trainLoader,inception),evaluation_inception(testLoader,inception)))
```

```python
BATCH_SIZE=32

squeezeNet=squeezeNet.to(device)
opt=optim.Adam(squeezeNet.parameters())
loss_fn = nn.CrossEntropyLoss()
epochs=2
n_iters = math.ceil(50000/BATCH_SIZE)

best_model_squeeze = fit(trainLoader,squeezeNet,opt,loss_fn,epochs)

squeezeNet.load_state_dict(best_model_squeeze)
print('Train accuracy: {}, Test accuracy: {}'.format(evaluation(trainLoader,squeezeNet),evaluation(testLoader,squeezeNet)))
```
