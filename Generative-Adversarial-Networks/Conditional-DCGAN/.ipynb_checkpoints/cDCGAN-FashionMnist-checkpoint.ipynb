```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.datasets import fashion_mnist

import warnings
warnings.filterwarnings('ignore')
```

### Loading , visualizing and preprocessing fashion-mnist dataset

```python
# loading dataset
(X_train, Y_train), (X_test,Y_test) = fashion_mnist.load_data()
print(X_train.shape,Y_train.shape)
```

```python
## concatinating test and train dataset (we don't need test dataset for GAN models)

X = np.concatenate((X_train,X_test))
Y = np.concatenate((Y_train,Y_test))

print(X.shape,Y.shape)
```

```python
## Visualizing images

def show_images(images,square_grid=True):
    if square_grid:
        plt.figure(figsize=(8,8))
        n = int(np.sqrt(images.shape[0]))
        for i in range(n*n):
            plt.subplot(n,n,i+1)
            plt.imshow(images[i],cmap='gray_r')
            plt.axis('off')
    else:
        for i in range(images.shape[0]):
            plt.figure(figsize=(0.8,0.8))
            plt.imshow(images[i],cmap='gray_r')
            plt.axis('off')
            plt.show()

show_images(X[:100])
```

```python
## expanding axis of the dataset

X = np.expand_dims(X,axis=-1)
Y = np.expand_dims(Y,axis=-1)

print(X.shape,Y.shape)
```

```python
## normalizing images to range [-1,1]

X = (X-127.5)/127.5
print(f"Max= {np.max(X)}")
print(f"Min= {np.min(X)}")
```

### Defining Discriminator model architecture

```python
def discriminator_model(input_shape=(28,28,1), num_classes=10, show_summary = False):
    in_image = Input(shape = input_shape)
    in_label = Input(shape= (1,))

    x1 = Embedding(num_classes,50)(in_label)
    x1 = Dense(input_shape[0]*input_shape[1])(x1)
    x1 = Reshape((input_shape[0],input_shape[1],1))(x1)

    x = Concatenate(axis=-1)([in_image,x1])

    x = Conv2D(filters=32,kernel_size=(3,3),padding='same',strides=2)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Conv2D(filters=64, kernel_size=(3,3),padding='same',strides=2)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Conv2D(filters=128, kernel_size=(3,3),padding='same',strides=2)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Flatten()(x)
    x = Dense(units=256)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.4)(x)
    x = Dense(units=64)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dense(units=1,activation='sigmoid')(x)
    
    model = tf.keras.Model([in_image,in_label],outputs=x)
    model.compile(optimizer = tf.keras.optimizers.Adam(lr=0.0002,beta_1=0.5), loss='binary_crossentropy',metrics=['accuracy'])
    
    if show_summary:
        print(model.summary())
        
    return model
```

### Defining Generator model architecture

```python
def generator_model(latent_dim=100,num_classes=10,show_summary = False):
    
    latent_input = Input(shape=(latent_dim,))
    in_label = Input(shape=(1,))
    
    x1 = Dense(7*7*128)(latent_input)
    x1 = LeakyReLU(0.2)(x1)
    x1 = Reshape((7,7,128))(x1)
    
    x2 = Embedding(num_classes,50)(in_label)
    x2 = Dense(49)(x2)
    x2 = Reshape((7,7,1))(x2)
    
    x = Concatenate(axis=-1)([x1,x2])
    x = Conv2DTranspose(64,kernel_size=(3,3),padding='same',strides=2)(x)
    x = LeakyReLU(0.2)(x)
    x = Conv2DTranspose(32,kernel_size=(5,5),padding='same',strides=2)(x)
    x = LeakyReLU(0.2)(x)
    x = Conv2DTranspose(1, kernel_size=(5,5),padding='same',activation='tanh')(x)
    
    model = tf.keras.Model(inputs=[latent_input,in_label], outputs = x)
    
    if show_summary:
        print(model.summary())
    
    return model
```

### Defining Conditional DCGAN model

```python
def conditional_gan_model(discriminator,generator,show_summary = False):
    
    discriminator.trainable=False

    generator_noise, generator_label = generator.input
    gan_output = discriminator([generator.output,generator_label])
    
    model = tf.keras.Model([generator_noise,generator_label], gan_output)
    
    model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.0002,beta_1=0.5),loss='binary_crossentropy')
    
    if show_summary:
        print(model.summary())
        
    return model
```

```python
def get_real_image_samples(X,Y,batch_size=64):
    
    idx = np.random.randint(0,X.shape[0],batch_size)
    images = X[idx]
    labels = Y[idx]
               
    return images,labels

def get_latent_data(latent_dim,batch_size):
    
    latent_input = np.random.randn(batch_size,latent_dim)
    labels = np.random.randint(0,10,(batch_size,1))
    
    return latent_input,labels

def get_fake_image_samples(generator,batch_size=64):

    latent_input, labels = get_latent_data(100,batch_size)
    
    images = generator([latent_input,labels])
    
    return images, labels
```

```python
def train_gan(X,Y,generator,discriminator,conditional_gan,epochs=10,batch_size=128):
    
    steps_per_epoch = X.shape[0]//batch_size
    half_batch = batch_size//2
    
    for epoch in range(epochs):
        print(f"Epoch #: {epoch}")
        
        for step in range(steps_per_epoch):
            
            ## training standalone discriminator
            real_images,real_labels = get_real_image_samples(X,Y,half_batch)
            fake_images,fake_labels = get_fake_image_samples(generator,half_batch)
            
            images = np.concatenate((real_images,fake_images))
            labels = np.concatenate((real_labels,fake_labels))
            dy = np.zeros((batch_size,1)) 
            dy[:half_batch]=1
            
            dloss, _ = discriminator.train_on_batch([images,labels],dy)
            
            ## training GAN
            
            generated_noise, generated_labels = get_latent_data(100,batch_size)
            gy = np.ones((batch_size,1)) 
            gy[:] = 0.9
            
            gloss = conditional_gan.train_on_batch([generated_noise,generated_labels],gy)
        
            if (step+1)%100==0:
                print(f"{step+1}/{steps_per_epoch} :: Generator loss: {gloss}  Discriminator loss:{dloss}")

    generator.save('generator_FashionMNIST.h5')
```

```python
# initializing models

generator = generator_model()
discriminator = discriminator_model()
conditional_gan = conditional_gan_model(discriminator,generator,show_summary=True)
```

```python
# calling train function to start training the GAN

train_gan(X, Y, generator, discriminator, conditional_gan, epochs=10, batch_size=256)
```

## Results

```python
cDCGAN_generator = tf.keras.models.load_model('generator_FashionMNIST.h5')

def generate_images(generator,num_images,label):
    
    latent_input,labels = get_latent_data(latent_dim=100,batch_size=num_images)
    labels[:,0]=label
    
    images = generator.predict([latent_input,labels])
    
    images = np.squeeze(images,axis=-1)
    
    if np.sqrt(num_images)-int(np.sqrt(num_images))==0.0:
        show_images(images,square_grid=True)
    else:
        show_images(images,square_grid=False)
```

```python
## generating images for all labels at once
label = np.asarray([x for _ in range(10) for x in range(10)])
generate_images(cDCGAN_generator,100,label)
```

```python
## generating images of specific label
label = 6
generate_images(cDCGAN_generator,10,label)
```
