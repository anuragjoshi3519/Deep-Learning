```python
# utility function to show images

def showImg(img,title='Image'):
    correcting_std = np.asarray([0.5,0.5,0.5]).reshape(3,1,1)
    correcting_mean = np.asarray([0.5,0.5,0.5]).reshape(3,1,1)
    
    img = img.numpy()
    img = np.multiply(img,correcting_std) + correcting_std
    img = np.transpose(img,(1,2,0))

    plt.axis('off')
    plt.title(title)
    plt.imshow(img)
```

# Occlusion Analysis 

```python
import seaborn
```

```python
# creating data loader
mobileNet.eval()
evalImages = torchvision.datasets.ImageFolder(root='Pokemon',transform=transform)
idx_to_classes = {id1:class1 for class1,id1 in evalImages.class_to_idx.items()}

imageLoader = torch.utils.data.dataloader.DataLoader(evalImages,batch_size=1,shuffle=True)
imageIter = iter(imageLoader)
```

```python
def occlusion(model,image,label,occ_size=50,occ_stride=50,occ_pixels=0.5):
    width,height = image.shape[-2],image.shape[-1]
    
    occ_width = math.ceil((width - occ_size)/occ_stride)
    occ_height = math.ceil((height - occ_size)/occ_stride)
    
    heatmap = torch.zeros((occ_width,occ_height))
    
    for h in range(0,height):
        for w in range(0,width):
            h_start = h*occ_stride
            w_start = w*occ_stride
            h_end = min(height,h_start+occ_size)
            w_end = min(width,w_start+occ_size)
            
            if w_end>=width or h_end>=height:
                continue
            
            input_image = image.clone().detach()
            input_image[:,:,w_start:w_end,h_start:h_end]=occ_pixels
            
            outputs = model(input_image)
            outputs = nn.functional.softmax(outputs,dim=1)
            prob = outputs.tolist()[0][label]
            
            heatmap[h,w] = prob 
    return heatmap        
```

```python
# image loader iterator
image,label = imageIter.next()
showImg(image[0],title = idx_to_classes[label.item()])

image=image.to(device)
label=label.to(device)

#finding prob before occlusion
outputs = mobileNet(image)
outputs = nn.functional.softmax(outputs,dim=1)
print(outputs.shape)
no_occ_pred,label = torch.max(outputs,1)
no_occ_pred = no_occ_pred.item()
print(no_occ_pred)
```

```python
# calling occlusion and creating heatmap
heatmap = occlusion(mobileNet,image,label.item(),occ_size=10,occ_stride=10)
img = seaborn.heatmap(heatmap, xticklabels=False,yticklabels=False,vmax=no_occ_pred)
```
